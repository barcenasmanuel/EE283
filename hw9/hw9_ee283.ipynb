```r
# Load necessary libraries
library(tidyverse)
library(nycflights13)
library(gridExtra)
library(grid)
```

```r
# Define temp_flights
temp_flights <- flights %>%
	group_by(carrier) %>%
	summarize(m_arr_delay = mean(arr_delay,na.rm=TRUE))

# P1: Scatter plot with a smoother
P1 <- ggplot(flights, aes(x = distance, y = arr_delay)) +
  geom_point(alpha = 0.3, size = 0.5) +
  geom_smooth(method = "loess", color = "blue") +
  labs(title = "Arrival Delay vs Distance", x = "Distance (miles)", y = "Arrival Delay (minutes)") +
  theme_minimal(base_size = 8)

# P2: Bar plot of mean arrival delay by carrier
P2 <- ggplot(temp_flights, aes(x = reorder(carrier, -m_arr_delay), y = m_arr_delay)) +
  geom_bar(stat = "identity", fill = "lightblue") +
  labs(title = "Mean Arrival Delay by Carrier", x = "Carrier", y = "Mean Arrival Delay (minutes)") +
  theme_minimal(base_size = 8) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# P3: Boxplot of arrival delay by carrier
P3 <- ggplot(flights, aes(x = carrier, y = arr_delay)) +
  geom_boxplot(outlier.size = 0.5) +
  labs(title = "Arrival Delay Distribution by Carrier", x = "Carrier", y = "Arrival Delay (minutes)") +
  theme_minimal(base_size = 8) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# P4: Histogram of arrival delays
P4 <- ggplot(flights, aes(x = arr_delay)) +
  geom_histogram(binwidth = 10, fill = "lightgreen", color = "black") +
  labs(title = "Distribution of Arrival Delays", x = "Arrival Delay (minutes)", y = "Frequency") +
  theme_minimal(base_size = 8)

# Define the layout matrix
lay <- rbind(c(1, 1, 1, 2),
             c(1, 1, 1, 3),
             c(1, 1, 1, 4))
```

```r
grid.arrange(P1,P2,P3,P4,ncol=2)

lay <- rbind(c(1,1,1,2),
             c(1,1,1,3),
             c(1,1,1,4))

tiff("figure1_modified.tiff", width = 7, height = 6, units = "in", res=600)
grid.arrange(P1,P2,P3,P4, layout_matrix = lay)
graphics.off()
```

# Problem 2

Make a 4 or 6 panel figure from some previous results.  A great one would be left-right paired fragments lengths and TSS enrichment plots for four ATACseq samples.

```r
# Load necessary libraries
library(gridExtra)
library(grid)
library(png)
```

```r
# Set the working directory
setwd("/pub/mbarcen1/EE283/ATACseq/output/ucsc/sub_processed")
```

```r
# Define the list of BAM files
bamfiles <- c("A4_ED_2.chrX.bam", "A4_WD_2.chrX.bam", 
              "A5_ED_3.chrX.bam", "A5_WD_3.chrX.bam", 
              "A6_ED_1.chrX.bam", "A6_WD_1.chrX.bam")

# Extract sample names from BAM files
sample_names <- gsub(".bam", "", bamfiles)

# Load images into lists
heatmaps <- lapply(sample_names, function(name) {
  readPNG(file.path("heatmap", paste0(name, "_heatmap.png")))
})

fragsizes <- lapply(sample_names, function(name) {
  readPNG(file.path("fragment_size_plots", paste0(name, "_fragsize.png")))
})
```

```r
# Create a list of grobs for the paired plots and their labels
paired_plots <- mapply(
  function(heatmap, fragsize, name) {
    list(
      textGrob(name, rot = 0, just = "left", gp = gpar(fontsize = 10, fontface = "bold")),
      arrangeGrob(rasterGrob(fragsize), rasterGrob(heatmap), ncol = 2)
    )
  },
  heatmaps, fragsizes, sample_names, SIMPLIFY = FALSE
)

# Flatten the list of paired plots
paired_plots <- do.call(c, paired_plots)

# Define a layout matrix
layout_matrix <- matrix(seq(1, length(paired_plots)), ncol = 2, byrow = TRUE)

# Arrange the plots in a grid
grid.arrange(
  grobs = paired_plots,
  layout_matrix = layout_matrix,
  widths = c(1,1),  # Adjust widths 
  top = textGrob("ATAC-seq Sample Analysis", gp = gpar(fontsize = 14, fontface = "bold"))
)
```

```r
# Read the results from the two CSV files
results_first_model <- read_csv("genome_analysis_results_new.csv")
results_second_model <- read_csv("genome_analysis_results_task2_new.csv")
```

```r
# Convert chromosome data to a factor and ensure it's in an appropriate format
results_first_model <- results_first_model %>%
  mutate(chr = as.factor(gsub("chr", "", chr))) # Remove 'chr' prefix and convert to factor

results_second_model <- results_second_model %>%
  mutate(chr = as.factor(gsub("chr", "", chr))) # Remove 'chr' prefix and convert to factor
```

```r
# Check if the SNP column exists
if(!"SNP" %in% names(results_first_model)) {
  results_first_model$SNP <- paste0("SNP", seq_along(results_first_model$chr))
}

# Similarly, check for the second model
if(!"SNP" %in% names(results_second_model)) {
  results_second_model$SNP <- paste0("SNP", seq_along(results_second_model$chr))
}
```

```r
# Define a function to convert complex chromosome identifiers to numeric
convert_chr_to_numeric <- function(chr) {
  chr_map <- c('2L' = 1, '2R' = 2, '3L' = 3, '3R' = 4, 'X' = 5)
  as.numeric(chr_map[chr])
}
```

```r
# Convert chromosome data to numeric using the map function
results_first_model <- results_first_model %>%
  mutate(chr = map_dbl(chr, convert_chr_to_numeric))

results_second_model <- results_second_model %>%
  mutate(chr = map_dbl(chr, convert_chr_to_numeric))
```

```r
# Summary of log10_p values to check for Inf or NA
summary(results_first_model$log10_p)

# Remove or impute NA/Inf values if necessary
results_first_model <- results_first_model %>%
  filter(is.finite(log10_p))
```

```r
# Calculate safe ylim values
ylim_values <- range(results_first_model$log10_p, na.rm = TRUE)
ylim_values[2] <- ylim_values[2] + 1  # Add a buffer to the upper limit

# Check the calculated ylim values
print(ylim_values)
```

```r
# Specify the file to save the plot
png("manhattan_plot_first_model.png", width = 1600, height = 1200, res = 300)

manhattan(results_first_model, chr="chr", bp="pos", p="log10_p",
            col=c("blue4", "orange3"), 
            main="Manhattan Plot for First Model",
           ylim=ylim_values)

dev.off()
```

```r
manhattan(results_first_model, chr="chr", bp="pos", p="log10_p",
            col=c("blue4", "orange3"), 
            main="Manhattan Plot for First Model",
           ylim=ylim_values)
```

```r
# Merge the datasets by SNP
results_combined <- inner_join(results_first_model, results_second_model, by = "SNP", suffix = c(".first", ".second"))

# Plotting the p-values
pvalueplot <- ggplot(results_combined, aes(x = P.first, y = P.second)) +
  geom_point(alpha = 0.6) +  # Using semi-transparent points
  theme_minimal() +
  labs(x = "P-value Model 1",
       y = "P-value Model 2",
       title = "Comparison of P-values between Two Models",
       subtitle = "Each point represents a SNP") +
  scale_x_log10(labels = scales::label_number()) +
  scale_y_log10(labels = scales::label_number()) +
  geom_abline(intercept = 0, slope = 1, linetype = "dashed", color = "red") # Adding a reference line y = x

pvalueplot
```
